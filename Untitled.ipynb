```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
import mediapipe as mp
import time
```

```python
from mtcnn import MTCNN  # MTCNN library needs to be installed
detector = MTCNN()
def detect_faces(frame):
    """Detect faces in the given video frame."""
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    faces = detector.detect_faces(rgb_frame)
    
    for face in faces:
        x, y, width, height = face['box']
        confidence = face['confidence']
        
        # Draw bounding box
        if confidence > 0.90:
            cv2.rectangle(frame, (x, y), (x + width, y + height), (0, 255, 0), 2)
            cv2.putText(frame, f"Confidence: {confidence:.2f}", (x, y - 10), 
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
    return frame
```

```python
# Initialize Mediapipe Face Mesh
mp_face_mesh = mp.solutions.face_mesh
face_mesh = mp_face_mesh.FaceMesh(refine_landmarks=True, min_detection_confidence=0.5, min_tracking_confidence=0.5)
```

```python
# Function to calculate eye aspect ratio (EAR)
def calculate_ear(landmarks, eye_indices):
    """Calculate Eye Aspect Ratio (EAR) to detect blinking."""
    # Extract the required landmarks for the eye
    p1, p2, p3, p4, p5, p6 = [landmarks[i] for i in eye_indices]
    # Calculate distances between points
    vertical_1 = np.linalg.norm(np.array([p2.x, p2.y]) - np.array([p6.x, p6.y]))
    vertical_2 = np.linalg.norm(np.array([p3.x, p3.y]) - np.array([p5.x, p5.y]))
    horizontal = np.linalg.norm(np.array([p1.x, p1.y]) - np.array([p4.x, p4.y]))
    # Compute EAR
    ear = (vertical_1 + vertical_2) / (2.0 * horizontal)
    return ear
```

```python
# Eye landmarks for Mediapipe (Right and Left eyes)
RIGHT_EYE = [33, 160, 158, 133, 153, 144]
LEFT_EYE = [362, 385, 387, 263, 373, 380]

# Thresholds
EAR_THRESHOLD = 0.2  # Threshold to detect blinking
BLINK_FRAMES = 2  # Consecutive frames for a blink
```

```python
# Start webcam feed
cap = cv2.VideoCapture(0)
print("Press 'q' to quit.")
blink_count = 0
consecutive_frames = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Flip and convert frame for Mediapipe
    frame = cv2.flip(frame, 1)
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Process frame with Mediapipe
    results = face_mesh.process(rgb_frame)
    if results.multi_face_landmarks:
        for face_landmarks in results.multi_face_landmarks:
            # Extract EAR for both eyes
            right_ear = calculate_ear(face_landmarks.landmark, RIGHT_EYE)
            left_ear = calculate_ear(face_landmarks.landmark, LEFT_EYE)
            avg_ear = (right_ear + left_ear) / 2.0

            # Blink detection logic
            if avg_ear < EAR_THRESHOLD:
                consecutive_frames += 1
            else:
                if consecutive_frames >= BLINK_FRAMES:
                    blink_count += 1
                    print(f"Blinked! Total blinks: {blink_count}")
                consecutive_frames = 0

            # Draw landmarks for eyes
            for eye in [RIGHT_EYE, LEFT_EYE]:
                for idx in eye:
                    x = int(face_landmarks.landmark[idx].x * frame.shape[1])
                    y = int(face_landmarks.landmark[idx].y * frame.shape[0])
                    cv2.circle(frame, (x, y), 2, (0, 255, 0), -1)

    # Display the video
    cv2.putText(frame, f"Blinks: {blink_count}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    cv2.imshow("Video KYC - Liveness Detection (Blinking)", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):  # Quit on 'q' key
        break

cap.release()
cv2.destroyAllWindows()
```

```python
# Initialize Mediapipe Face Mesh and Drawing utils
mp_face_mesh = mp.solutions.face_mesh
mp_drawing = mp.solutions.drawing_utils
mp_drawing_styles = mp.solutions.drawing_styles
face_mesh = mp_face_mesh.FaceMesh(refine_landmarks=True, min_detection_confidence=0.5, min_tracking_confidence=0.5)

# 3D Model points of the head (based on Mediapipe landmarks)
FACE_3D_POINTS = np.array([
    (0, 0, 0),    # Nose tip
    (0, -330, -65),  # Chin
    (-225, 170, -135),  # Left eye left corner
    (225, 170, -135),   # Right eye right corner
    (-150, -150, -125), # Left Mouth corner
    (150, -150, -125)   # Right Mouth corner
], dtype=np.float32)

# Corresponding 2D points for projection
FACE_2D_LANDMARKS = [
    1, 152, 33, 263, 61, 291  # Nose tip, Chin, Left eye, Right eye, Left mouth, Right mouth
]

# Camera parameters (Assume a generic camera for now)
CAMERA_MATRIX = np.array([[640, 0, 320],
                          [0, 640, 240],
                          [0, 0, 1]], dtype=np.float32)
DIST_COEFFS = np.zeros((4, 1))  # No lens distortion

# Start webcam
cap = cv2.VideoCapture(0)
print("Press 'q' to quit.")

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Flip and convert to RGB
    frame = cv2.flip(frame, 1)
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Process with Mediapipe Face Mesh
    results = face_mesh.process(rgb_frame)

    if results.multi_face_landmarks:
        for face_landmarks in results.multi_face_landmarks:
            # Extract 2D landmarks
            landmarks = np.array([(int(lm.x * frame.shape[1]), int(lm.y * frame.shape[0])) for lm in face_landmarks.landmark])
            face_2d_points = landmarks[FACE_2D_LANDMARKS, :]

            # SolvePnP to get head pose
            success, rotation_vector, translation_vector = cv2.solvePnP(FACE_3D_POINTS, face_2d_points, CAMERA_MATRIX, DIST_COEFFS)
            
            # Get rotation angles (yaw, pitch, roll)
            rotation_matrix, _ = cv2.Rodrigues(rotation_vector)
            angles, _, _, _, _, _ = cv2.RQDecomp3x3(rotation_matrix)
            yaw, pitch, roll = angles

            # Draw head pose info
            cv2.putText(frame, f"Yaw: {yaw:.2f}", (20, 40), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)
            cv2.putText(frame, f"Pitch: {pitch:.2f}", (20, 60), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)
            cv2.putText(frame, f"Roll: {roll:.2f}", (20, 80), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

            # Visualize landmarks
            mp_drawing.draw_landmarks(
                frame,
                face_landmarks,
                mp_face_mesh.FACEMESH_CONTOURS,
                landmark_drawing_spec=mp_drawing_styles.get_default_face_mesh_contours_style()
            )

    # Display the video
    cv2.imshow("Video KYC - Head Movement Detection", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):  # Quit on 'q' key
        break

cap.release()
cv2.destroyAllWindows()
```

```python
embedder = FaceNet()
detector = MTCNN()
if not os.path.exists('captured_photos'):
    os.makedirs('captured_photos')
def capture_live_photo():
    """Capture live photo from webcam if face is detected."""
    cap = cv2.VideoCapture(0)
    while True:
        ret, frame = cap.read()
        if not ret:
            print("Failed to grab frame.")
            break
        faces = detect_faces(frame)
        if len(faces) > 0:
            # If faces detected, save the first face as live photo
            cv2.imwrite('captured_photos/live_photo.jpg', frame)  # Save as live photo
            print("Live photo captured!")
            break
        cv2.imshow('Webcam', frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    cap.release()
    cv2.destroyAllWindows()
def capture_id_photo():
    """Continuously capture frames until an ID photo with a face is detected."""
    cap = cv2.VideoCapture(0)
    while True:
        ret, frame = cap.read()
        if not ret:
            print("Failed to grab frame.")
            break
        faces = detect_faces(frame)
        if len(faces) > 0:
            cv2.imwrite('captured_photos/id_photo.jpg', frame)  # Save as ID photo
            print("ID photo captured!")
            break  # Exit the loop once the photo is captured
        else:
            cv2.putText(frame, "No face detected. Adjust ID photo.", (50, 50), 
                        cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
            cv2.imshow('Webcam', frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    cap.release()
    cv2.destroyAllWindows()
def detect_faces(frame):
    """Detect faces in the given image using MTCNN."""
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)  # Convert image to RGB
    faces = detector.detect_faces(rgb_frame)  # Detect faces
    face_images = []
    for face in faces:
        x, y, width, height = face['box']
        confidence = face['confidence']
        if confidence > 0.90:  # Use faces with high confidence
            cropped_face = frame[y:y + height, x:x + width]  # Crop face region
            cropped_face = cv2.resize(cropped_face, (160, 160))  # Resize to 160x160
            cropped_face = cv2.cvtColor(cropped_face, cv2.COLOR_BGR2RGB)  # Convert to RGB
            face_images.append(cropped_face)
    return face_images
def show_image(image, title="Image"):
    """Display the image with Matplotlib."""
    plt.imshow(image)
    plt.title(title)
    plt.axis('off')
    plt.show()
def calculate_similarity(embedding1, embedding2):
    """Calculate cosine similarity between two face embeddings."""
    from numpy.linalg import norm
    dot_product = np.dot(embedding1, embedding2)
    norm1 = norm(embedding1)
    norm2 = norm(embedding2)
    return dot_product / (norm1 * norm2)
```

```python
import cv2
import numpy as np
from mtcnn import MTCNN
import time
import os

# Initialize FaceNet model and MTCNN face detector
detector = MTCNN()

# Create a directory to save the images if it doesn't exist
if not os.path.exists('captured_photos'):
    os.makedirs('captured_photos')

def detect_faces(frame):
    """Detect faces in the given image using MTCNN."""
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)  # Convert image to RGB
    faces = detector.detect_faces(rgb_frame)  # Detect faces
    
    face_images = []
    for face in faces:
        x, y, width, height = face['box']
        confidence = face['confidence']
        
        if confidence > 0.90:  # Use faces with high confidence
            cropped_face = frame[y:y + height, x:x + width]  # Crop face region
            cropped_face = cv2.resize(cropped_face, (160, 160))  # Resize to 160x160
            cropped_face = cv2.cvtColor(cropped_face, cv2.COLOR_BGR2RGB)  # Convert to RGB
            face_images.append(cropped_face)
    
    return face_images

def improve_image_quality(frame):
    """Improve the quality of the captured frame for better embeddings."""
    # Increase the resolution (optional based on the camera capability)
    # Here we resize the frame to 1080p (1920x1080) resolution for more detail
    frame = cv2.resize(frame, (1920, 1080))

    # Histogram Equalization (Improves contrast)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    equalized = cv2.equalizeHist(gray)
    frame = cv2.cvtColor(equalized, cv2.COLOR_GRAY2BGR)

    # Optional: Increase brightness/contrast (if needed)
    # Apply some adjustments to make the face more visible
    alpha = 1.2  # Contrast control
    beta = 20    # Brightness control
    frame = cv2.convertScaleAbs(frame, alpha=alpha, beta=beta)
    
    return frame

def capture_id_photo():
    """Continuously capture frames until an ID photo with a face is detected."""
    # Open webcam
    cap = cv2.VideoCapture(0)
    
    while True:
        ret, frame = cap.read()
        if not ret:
            print("Failed to grab frame.")
            break
        
        # Improve the quality of the frame
        frame = improve_image_quality(frame)
        
        # Detect faces in the improved frame
        faces = detect_faces(frame)
        
        if len(faces) > 0:
            # If faces are detected, save the first face as ID photo
            cv2.imwrite('captured_photos/id_photo.jpg', frame)  # Save as ID photo
            print("ID photo captured!")
            break  # Exit the loop once the photo is captured
        else:
            # Display the frame with message if no face detected
            cv2.putText(frame, "No face detected. Adjust ID photo.", (50, 50), 
                        cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
            cv2.imshow('Webcam', frame)
        
        # Exit if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    
    # Release webcam and close window
    cap.release()
    cv2.destroyAllWindows()
```
